```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/thearolskovsloth/data-science-x-am/output/body_descriptions-w-lemmas.csv")
df_raw = pd.read_csv("/Users/thearolskovsloth/data-science-x-am/output/body_descriptions.csv")

df.head(20)
```

```python
#lowercase all body parts
df["bodypart"] = df["bodypart"].str.lower()
```

```python
#male authors top 50 words in a list
m_authors = df.loc[(df['author_gender']=="M")]
top20_m_authors = m_authors['bodypart'].value_counts()[:20].rename_axis('bodypart').reset_index(name='top20_m:authors')
top20_m_authors = list(top20_m_authors['bodypart'])
len(top20_m_authors)
```

```python
#female authors top 50 words in a list
f_authors = df.loc[(df['author_gender']=="F")]
top20_f_authors = f_authors['bodypart'].value_counts()[:20].rename_axis('bodypart').reset_index(name='top20_f:authors')
top20_f_authors = list(top20_f_authors['bodypart'])
len(top20_f_authors)
```

```python
male_owners_female_authors = df.loc[(df['owner_gender']=="M") & (df['author_gender']=="F")]
male_owners_male_authors = df.loc[(df['owner_gender']=="M") & (df['author_gender']=="M")]
print("male_owners_female_authors:",len(male_owners_female_authors),"male_owners_male_authors:", len(male_owners_male_authors))
male_owners_male_authors.head()
```

```python
#females
female_owners_female_authors = df.loc[(df['owner_gender']=="F") & (df['author_gender']=="F")]
female_owners_male_authors = df.loc[(df['owner_gender']=="F") & (df['author_gender']=="M")]
print("female_owners_female_authors:",len(female_owners_female_authors),"female_owners_male_authors:", len(female_owners_male_authors))
female_owners_female_authors.head()
```

```python
print(len(male_owners_female_authors.iloc[1,4]))
```

## female authors and bodiez

```python
#male_owners
male_owners_female_authors_counts = male_owners_female_authors['bodypart'].value_counts().rename_axis('bodypart').reset_index(name='counts_male_owners_female_authors')
#female_owners
female_owners_female_authors_counts = female_owners_female_authors['bodypart'].value_counts().rename_axis('bodypart').reset_index(name='counts_female_owners_female_authors')

male_owners_female_authors_counts = male_owners_female_authors_counts[male_owners_female_authors_counts['bodypart'].isin(top20_f_authors)]
female_owners_female_authors_counts = female_owners_female_authors_counts[female_owners_female_authors_counts['bodypart'].isin(top20_f_authors)]
len(female_owners_female_authors_counts)
```

## male authors  and bodiez
## 

```python
#male_owners
male_owners_male_authors_counts = male_owners_male_authors['bodypart'].value_counts().rename_axis('bodypart').reset_index(name='counts_male_owners_male_authors')
#female_owners
female_owners_male_authors_counts = female_owners_male_authors['bodypart'].value_counts().rename_axis('bodypart').reset_index(name='counts_female_owners_male_authors')

male_owners_male_authors_counts = male_owners_male_authors_counts[male_owners_male_authors_counts['bodypart'].isin(top20_m_authors)]
female_owners_male_authors_counts = female_owners_male_authors_counts[female_owners_male_authors_counts['bodypart'].isin(top20_m_authors)]

len(male_owners_male_authors_counts)
```

### standardized measures

```python
#female authors
male_owners_female_authors_counts['std_male_owners_female_authors'] = male_owners_female_authors_counts.apply (lambda row: (row.counts_male_owners_female_authors)/(male_owners_female_authors_counts['counts_male_owners_female_authors'].sum()), axis=1)

female_owners_female_authors_counts['std_female_owners_female_authors'] = female_owners_female_authors_counts.apply (lambda row: (row.counts_female_owners_female_authors)/(female_owners_female_authors_counts['counts_female_owners_female_authors'].sum()), axis=1)

print('total number of female bodybarts mentioned by female authors:', female_owners_female_authors_counts['counts_female_owners_female_authors'].sum(), '. total number of male bodybarts mentioned by female authors:', male_owners_female_authors_counts['counts_male_owners_female_authors'].sum() )
```

```python
#male authors
male_owners_male_authors_counts['std_male_owners_male_authors'] = male_owners_male_authors_counts.apply (lambda row: (row.counts_male_owners_male_authors)/(male_owners_male_authors_counts['counts_male_owners_male_authors'].sum()), axis=1)

female_owners_male_authors_counts['std_female_owners_male_authors'] = female_owners_male_authors_counts.apply (lambda row: (row.counts_female_owners_male_authors)/(female_owners_male_authors_counts['counts_female_owners_male_authors'].sum()), axis=1)

print('total number of female bodybarts mentioned by male authors:', female_owners_male_authors_counts['counts_female_owners_male_authors'].sum(), '. total number of male bodybarts mentioned by male authors:', male_owners_male_authors_counts['counts_male_owners_male_authors'].sum() )
```

### merge into female author df and male author df

```python
female_authors = pd.merge(male_owners_female_authors_counts, female_owners_female_authors_counts, on=['bodypart'])
female_authors = female_authors.drop(['counts_male_owners_female_authors', 'counts_female_owners_female_authors'], axis=1)
female_authors.tail()
```

```python
male_authors = pd.merge(male_owners_male_authors_counts, female_owners_male_authors_counts, on=['bodypart'])
male_authors = male_authors.drop(['counts_male_owners_male_authors', 'counts_female_owners_male_authors'], axis=1)
male_authors.tail()
```

```python
male_authors['skewness_male_authors'] = np.where(male_authors['std_male_owners_male_authors'] > male_authors['std_female_owners_male_authors'], ((male_authors['std_male_owners_male_authors'])/(male_authors['std_female_owners_male_authors'])), ((-male_authors['std_female_owners_male_authors'])/male_authors['std_male_owners_male_authors']))

male_authors['n'] = range(len(male_authors))
male_authors['skew_direction'] = np.where(male_authors['skewness_male_authors'] > 0, "male_skew", "female_skew")

male_authors['abs_skew_male_authors'] = abs(male_authors['skewness_male_authors'])

male_authors.tail()
```

```python
#prepare file
word_list_female = []
word_list_male = []
def repeat(word, n):
    
    repeat_word = word[:]
    result = ""

    for i in range(n):
        result = result + " " + repeat_word

    word_list_male.append(result)

for w in range(1,len(top100females)):
    stripped_w = re.sub(r'[^a-zA-Z]', '', str(top100females.iloc[w,0]))
    repeat(stripped_w, top100females.iloc[w,1])

words_male=' '.join(map(str,word_list_male))
words_female=' '.join(map(str,word_list_female))
```
